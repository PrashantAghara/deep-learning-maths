```python
import numpy as np
```

#### Ways to create arrays in Numpy

```python
arr = np.array([10, 12])
print(arr)
```

```python
arr2 = np.arange(3)
print(arr2)
```

```python
arr3 = np.arange(1, 20, 3)
print(arr3)
```

```python
arr4 = np.linspace(0, 100, 5)
print(arr4)

arr4_int = np.linspace(0, 100, 5, dtype=int)
print(arr4_int)
```

```python
char_arr = np.array("Prashant Aghara")
print(char_arr)
print(char_arr.dtype)
```

```python
print(np.ones(3))
print(np.zeros(3))
print(np.empty(3))
print(np.random.rand(3))
```

#### 2D Array

```python
two_d_arr = np.array([[1,2,3],[4,5,6]])
print(two_d_arr)
```

```python
one_dim_arr = np.array([1,2,3,4,5,6])

multi_dim_arr = np.reshape(one_dim_arr, (2,3))
print(multi_dim_arr)
```
