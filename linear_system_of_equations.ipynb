#### Basic Linear System operations

```python
A = np.array([
        [-1, 3],
        [3, 2]
    ], dtype=np.dtype(float))

b = np.array([7, 1], dtype=np.dtype(float))

print("Matrix A:")
print(A)
print("\nArray b:")
print(b)
```

```python
x = np.linalg.solve(A, b)

print(f"Solution: {x}")
```

```python
print(np.linalg.det(A))
```
